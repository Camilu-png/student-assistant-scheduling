```python
import pandas as pd

total_285 = 31
total_295 = 26
```

# Settings for experiment 1

```python
df285 = pd.read_csv("../../results/configurations/experiment1-INF285.csv")
df295 = pd.read_csv("../../results/configurations/experiment1-INF295.csv")
```

```python
print(df285.head())
print(df295.head())
```

```python
valid_df285 = df285[df285["validity"]].copy()
valid_df295 = df295[df295["validity"]].copy()

best_config_285 = valid_df285.sort_values(
    by=["final_fitness", "time"], ascending=[False, True]
).iloc[0]

best_config_295 = valid_df295.sort_values(
    by=["final_fitness", "time"], ascending=[False, True]
).iloc[0]
```

```python
print("Best configuration for INF-285:")
print(best_config_285)
print(
    f"Standardized fitness: {round(best_config_285['final_fitness'] * 100 / total_285, 2)}\n"
)
print("Best configuration for INF-295:")
print(best_config_295)
print(
    f"Standardized fitness: {round(best_config_295['final_fitness'] * 100 / total_295, 2)}\n"
)
```

```python
print("Average time per configuration for INF-285:")
print(valid_df285["time"].mean().round(2))

print("Average time per configuration for INF-295:")
print(valid_df295["time"].mean().round(2))
```
